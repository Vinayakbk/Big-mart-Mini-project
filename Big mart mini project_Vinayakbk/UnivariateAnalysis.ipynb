```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('Train.csv')
fig=plt.figure()
sns.scatterplot(data.index, data['Item_Visibility'])
plt.show()
```

```python
sns.set(rc={'figure.figsize':(7,7)})
sns.set(font_scale=1.5)
fig=sns.lineplot(x=data.index,y=data['Item_Weight'],markevery=1,marker='d',data=data,hue=data['Item_MRP'])
fig.set(xlabel='Item_Type')
```

```python
sns.stripplot(x=data['Item_Weight'], y=data['Item_Type'])
```

```python
plt.hist(data['Item_Type'])
```

```python
plt.figure(figsize=(5,5))
data['Item_MRP'].plot(kind='density')
```

```python
fig, ax = plt.subplots()
sns.rugplot(data['Item_Weight'])
ax.set_xlim(3,9)
plt.show()
```

```python
sns.distplot(data['Item_Weight'],rug=True,hist=False)
```

```python
plt.boxplot(data['Item_MRP'])
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.distplot(data['Item_Weight'], color='black',rug=True)
```

```python
plt.figure(figsize=(5,5))
plt.pie(data['Outlet_Location_Type'].value_counts(),startangle=90,autopct='%.3f',
        labels=['Item_MRP','Item_Weight','Item_Visibility'],shadow=True)
```
